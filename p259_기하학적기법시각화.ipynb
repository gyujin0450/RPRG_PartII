```
# 한개 변수 대상으로 gplot() 함수 적용
# qplot(x축 ~ y축, data, facets, geom, stat, position, xlim, ylim, log, main, xlab, ylab, asp)
```

```
qplot(hwy, data= mpg)
```

```
qplot(hwy, data = mpg, fill = drv) # fill 속성을 적용하여 해당 범주에 서로 다른 색을 채움
```

```
qplot(hwy, data=mpg, fill = drv, binwidth = 1)
```

```
# facets 를 사용하여, drv 값으로 패널 생성(열)
qplot(hwy, data = mpg, fill = drv, facets = .~drv, binwidth = 2)
```

```
# facets 를 사용하여, drv 값으로 패널 생성(행)
qplot(hwy, data = mpg, fill = drv, facets = drv~., binwidth = 2)
```

```
# 두개 변수 대상으로 qplot() 함수 적용
# 1) 두변수로 disl과 hwy 변수 사용, drv 변수로 색상 적용
qplot(displ, hwy, data = mpg, color = drv)
```

```
# drv 변수로 패널 구분(열)
qplot(displ, hwy, data = mpg, color = drv, facets = .~drv)
```
