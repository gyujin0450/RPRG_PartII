```
# qplot() 함수 : 점,선,다각형 등의 기하학적 객체를 크기, 모양, 색상 등의 미적 요소를 매핑하여 그래프를 그려줌
install.packages("ggplot2")
library(ggplot2)
```

```
# ?mpg
# the fuel economy data , It contains only models which had a new release every year between 1999 and  2008 
# manufacturer : manufacturer name
# model : model name
# displ : engine displacement(배기량), in litres
# year  : year of manufacture
# cyl   : number of cylinders
# trans : type of transmission(변속기)
# drv   : the type of drive train(구동), where f = front-wheel drive, r = rear wheel drive, 4 = 4wd
# cty   : city miles per gallon
# hwy   : highway miles per gallon
# fl    : fuel type
# class : "type" of car
```

```
# 한개 변수 대상으로 gplot() 함수 적용
# qplot(x축 ~ y축, data, facets, geom, stat, position, xlim, ylim, log, main, xlab, ylab, asp)
```

```
qplot(hwy, data= mpg)
```

```
qplot(hwy, data = mpg, fill = drv) # fill 속성을 적용하여 해당 범주에 서로 다른 색을 채움
```

```
qplot(hwy, data=mpg, fill = drv, binwidth = 1)
```

```
# facets 를 사용하여, drv 값으로 패널 생성(열)
qplot(hwy, data = mpg, fill = drv, facets = .~drv, binwidth = 2)
```

```
# facets 를 사용하여, drv 값으로 패널 생성(행)
qplot(hwy, data = mpg, fill = drv, facets = drv~., binwidth = 2)
```

```
# 두개 변수 대상으로 qplot() 함수 적용
# 1) 두변수로 disl과 hwy 변수 사용, drv 변수로 색상 적용
qplot(displ, hwy, data = mpg, color = drv)
```

```
# drv 변수로 패널 구분(열)
qplot(displ, hwy, data = mpg, color = drv, facets = .~drv)
```

```
# 미적 요소 매핑 : 색상, 크기, 모양 등의 미적 요소를 데이터에 연결함
str(mtcars)
# ?mtcars
# The data was extracted from the 1974 _Motor Trend_ US magazine,
# and comprises fuel consumption and 10 aspects of automobile design
# and performance for 32 automobiles (1973-74 models).
# mpg   Miles/(US) gallon                        
# cyl   Number of cylinders                      
# disp  Displacement (cu.in.)                    
# hp    Gross horsepower                         
# drat  Rear axle ratio                          
# wt    Weight (1000 lbs)                        
# qsec  1/4 mile time                            
# vs    Engine (0 = V-shaped, 1 = straight)      
# am    Transmission (0 = automatic, 1 = manual) 
# gear  Number of forward gears                  
# carb  Number of carburetors                          
```

```
head(mtcars)
```

```
# 색상, 크기 및 모양 적용
qplot(wt, mpg, data = mtcars, color = factor(carb), size = qsec, shape = factor(cyl))
```

```
# 기하학적 객체 적용
str(diamonds)
# ?diamonds
# A dataset containing the prices and other attributes of almost 54,000 diamonds.
# carat : weight of the diamond (0.2-5.01) 
# cut : quality of the cut (Fair, Good, Very Good, Premium, Ideal)
# color : diamond colour, from D (best) to J (worst)
# clarity : a measurement of how clear the diamond is (I1 (worst),
#           SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
# depth : total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43-79)
# table : width of top of diamond relative to widest point (43-95)
# price : price in US dollars (\$326-\$18,823)
# x : length in mm (0-10.74)
# y : width in mm (0-58.9)
# z : depth in mm (0-31.8)
```

```
 head(diamonds)
```

```
summary(diamonds)
```

```
# geom 속성과 fill 속성
qplot(clarity, data = diamonds, fill = cut, geom = "bar" )
```

```
qplot(clarity, data = diamonds, color = cut, geom = "bar")
```

```
qplot(wt,mpg, data = mtcars, size = qsec, geom = "point")
```

```
qplot(wt,mpg, data = mtcars, size = qsec, geom = "point",
      color = factor(carb), shape = factor(cyl))
```

```
# geom = "smooth" 속성으로 산점도 그래프의 평활 그리기
qplot(wt, mpg, data = mtcars, geom = "smooth")
```

```
qplot(wt, mpg, data = mtcars, geom = c("point","smooth")) # 해석? 
```

```
qplot(wt, mpg, data = mtcars, geom = c("line","point"), color = factor(cyl))
```
