```
## 히스토그램 시각화
# iris 데이터 셋 가져오기
data(iris)
names(iris)
```

```
str(iris)
```

```
head(iris)
```

```
summary(iris$Sepal.Length)
```

```
hist(iris$Sepal.Length, xlab = "iris$Sepal.Length", col = "magenta",
     main = "Sepal Length", xlim = c(4.3,7.9))
```

```
summary(iris$Sepal.Width)
```

```
hist(iris$Sepal.Width, xlab = "iris$Sepal.Width", col = "mistyrose",
     main = "Sepal Width", xlim = c(2.0,4.5))
```

```
## 히스토그램에서 빈도와 밀도 표현하기
par(mfrow = c(1,2))
# 빈도수에 의한 히스토그램 : [freq = T]
hist(iris$Sepal.Width, xlab = "iris$Sepal.Width", col = "green",
     main = "Sepal Width : frequency", xlim = c(2.0,4.5))
# 확률 밀도에 의한 히스토그램 : freq = F     
hist(iris$Sepal.Width, xlab = "iris$Sepal.Width", col = "mistyrose", freq = F,
     main = "Sepal Width : Density", xlim = c(2.0,4.5))
# 밀도를 기준으로 line 추가
lines(density(iris$Sepal.Width), col = "red")     
```

```
## 정규분포 추정 곡선
# 계급을 밀도로 표현한 히스토그램 시각화
par(mfrow = c(1,1))
hist(iris$Sepal.Width, xlab = "iris$Sepal.Width", col="mistyrose", freq=F,
     main = "Sepal Width : Density", xlim = c(2.0, 4.5))
lines(density(iris$Sepal.Width), col = "red") 
# 히스토그램에 정규분포 추정 곡선 추가
x <- seq(2.0, 4.5, 0.1)
curve(dnorm(x,mean = mean(iris$Sepal.Width),sd = sd(iris$Sepal.Width)),col="blue", add=T)    
```

```
## 산점도 시각화
# 기본 산점도 시각화
price <- runif(10, min = 1, max = 100)
plot(price, col = "red")
# 대각선 추가
par(new = T)
line_chart = 1:100
plot(line_chart, type = "l", col = "red", axes = F, ann = F)
# 텍스트 추가
text(70,80,"add test", col="blue")
```

```
# type 속성으로 산점도 그리기
round(price)
par(mfrow = c(2,2))
plot(price, type = "l") # 실선
plot(price, type = "o") # 원형과 실선
plot(price, type = "h") # 직선
plot(price, type = "s") # 꺽은선
```

```
# pch 속성으로 산점도
par(mfrow = c(2,2))
plot(price, type = "o", pch = 5) # 빈 사각형 
plot(price, type = "o", pch = 15) # 채원진 사각형 
plot(price, type = "o", pch = 20, col = "blue") # 파랑색 
plot(price, type = "o", pch = 20, col = "orange", cex = 3) #  cex : 점크기
```

```
par(mfrow = c(1,1))
# lwd : line width(선 굵기)
plot(price, type = "o", pch = 20, col = "orange", cex = 2, lwd = 3)
```

```
## galton 데이터 셋을 대상으로 중복된 자료 시각화
# galton 데이터셋 가져오기
install.packages("UsingR")
library(UsingR)
data(galton)
```

```
# 교차테이블을 작성하고, 데이터프레임으로 변환
galtonData <- as.data.frame(table(galton$child, galton$parent))
head(galtonData)
```

```
# 컬럼 단위 추출
names(galtonData) = c("child","parent","freq")
head(galtonData)
```

```
summary(galtonData)
```

```
plot(table(galtonData$freq),type = "o", pch = 20, col = "orange", cex = 3)
```

```
parent <- as.numeric(galtonData$parent)
child <- as.numeric(galtonData$child)
```

```
table(parent)
```

```
table(child)
```

```
plot(parent, child, pch = 21, col = "blue", bg = "green", cex = 1, xlab="parent", ylab="child")
```

```
## 그래프가 이해되지 않음
plot(parent, child, pch = 21, col = "blue", bg = "green",
     cex = 0.2 * galtonData$freq, xlab="parent", ylab="child")
```

```
## 3차원으로 산점도 시각화
install.packages("scatterplot3d") 
library(scatterplot3d)
```

```
# Factor의 levels 보기
iris_setosa = iris[iris$Species == 'setosa',]
iris_versicolor = iris[iris$Species == 'versicolor',]
iris_virginica = iris[iris$Species == 'virginica',]
```

```
d3 <- scatterplot3d(iris$Petal.Length, iris$Sepal.Length, iris$Sepal.Width, type = 'n')
d3$points3d(iris_setosa$Petal.Length, iris_setosa$Sepal.Length, iris_setosa$Sepal.Width, bg="orange", pch=21)
d3$points3d(iris_versicolor$Petal.Length, iris_versicolor$Sepal.Length, iris_versicolor$Sepal.Width,bg="blue", pch=23)
d3$points3d(iris_virginica$Petal.Length, iris_virginica$Sepal.Length, iris_virginica$Sepal.Width, bg="green", pch=25)
```
